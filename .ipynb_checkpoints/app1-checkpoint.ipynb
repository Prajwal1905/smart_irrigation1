```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report
import joblib
```

```python
df = pd.read_csv("irrigation.csv")
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
```

```python
print(df.info())
print(df.describe())
```

```python
X = df.iloc[:, :20]
y = df.iloc[:, 20:]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=4,
    min_samples_leaf=2,
    max_features='sqrt',
    random_state=42
)
```

```python
model = MultiOutputClassifier(rf)
model.fit(X_train, y_train)
```

```python
joblib.dump(model, "Farm_Irrigation_Model.pkl")
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=y.columns))
```

```python
plt.figure(figsize=(8, 4))
df[['parcel_0', 'parcel_1', 'parcel_2']].sum().plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title("Parcel Activation Counts")
plt.ylabel("Number of Activations")
plt.xlabel("Parcel")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
conditions = {
    "Parcel 0 ON": df['parcel_0'],
    "Parcel 1 ON": df['parcel_1'],
    "Parcel 2 ON": df['parcel_2'],
    "Parcel 0 & 1 ON": df['parcel_0'] & df['parcel_1'],
    "Parcel 0 & 2 ON": df['parcel_0'] & df['parcel_2'],
    "Parcel 1 & 2 ON": df['parcel_1'] & df['parcel_2'],
    "All Parcels ON": df['parcel_0'] & df['parcel_1'] & df['parcel_2'],
}
fig, axs = plt.subplots(nrows=len(conditions), figsize=(12, 18), sharex=True)
for ax, (title, condition) in zip(axs, conditions.items()):
    ax.step(df.index, condition.astype(int), where='post', linewidth=1.5, color='teal')
    ax.set_title(title)
    ax.set_ylabel("Status")
    ax.set_yticks([0, 1])
    ax.set_yticklabels(['OFF', 'ON'])
axs[-1].set_xlabel("Time Index")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.step(df.index, df['parcel_0'], where='post', label='Parcel 0', color='blue', linewidth=1.5)
plt.step(df.index, df['parcel_1'], where='post', label='Parcel 1', color='orange', linewidth=1.5)
plt.step(df.index, df['parcel_2'], where='post', label='Parcel 2', color='green', linewidth=1.5)

plt.title("Sprinkler Pump Activity Timeline")
plt.xlabel("Time Index")
plt.ylabel("Status (0=OFF, 1=ON)")
plt.yticks([0, 1], ['OFF', 'ON'])
plt.legend(loc='upper right')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
